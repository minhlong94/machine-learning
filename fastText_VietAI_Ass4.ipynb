# Use Facebook's fastText for sentiment analysis

```python
import pandas as pd
```

```python
pd.options.display.max_colwidth=1000
train_df = pd.read_csv("../input/vietai-dataset/assignment4-data/Assignment4/train.csv")
train_df.tail()
```

```python
import re
# re = regular expressions
strip_special_chars = re.compile("[^\w0-9 ]+")

def clean_sentences(string):
    string = string.lower().replace("<br />", " ")
    return re.sub(strip_special_chars, "", string.lower())
```

```python
cleaned_sentence = []
for idx, row in train_df.iterrows():
    cleaned_sentence.append(clean_sentences(row["text"]).replace("_", " "))
train_df["text"] = cleaned_sentence
```

```python
train_df
```

```python
txt_file = open("train.txt", "w+")
for idx, row in train_df.iterrows():
    txt_file.write("__label__{} {}\n".format("postive" if row["class"] == 1 else "negative", row["text"]))
```

```python
test_df = pd.read_csv("../input/vietai-dataset/assignment4-data/Assignment4/test.csv")
print('Number of test samples in total:', len(test_df))
test_df.head()
```

```python
import fasttext as ft
model = ft.train_supervised("./train.txt")
```

```python
# print(tokenized[0])
```

```python
result = []
for idx, row in test_df.iterrows():
    pred = model.predict(clean_sentences(row["text"]))
    if pred[0][0] == '__label__negative':
        result.append(0)
    else:
        result.append(1)
```

```python
test_df["class"] = result
test_df
```
